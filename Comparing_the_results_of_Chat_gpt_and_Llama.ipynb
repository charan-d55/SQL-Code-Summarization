Compairing the rouge-L and cosine similarity of both the models.

```python
import numpy as np
import matplotlib.pyplot as plt

# Example scores for LLAMA and ChatGPT (replace with your actual data)
llama_scores = [0.317,0.643]  # ROUGE, Cosine Similarity
chatgpt_scores = [0.426,0.712]  # ROUGE, Cosine Similarity

# Set the width of the bars
bar_width = 0.35

# Set the position of the bars on the x-axis
r1 = np.arange(len(llama_scores))
r2 = [x + bar_width for x in r1]

# Plot bars
plt.figure(figsize=(8, 6))
plt.bar(r1, llama_scores, color='teal', width=bar_width, edgecolor='grey', label='LLAMA')
plt.bar(r2, chatgpt_scores, color='orange', width=bar_width, edgecolor='grey', label='ChatGPT')

# Add xticks on the middle of the group bars
plt.xlabel('Metric', fontweight='bold')
plt.xticks([r + bar_width / 2 for r in range(len(llama_scores))], ['ROUGE-L', 'Cosine Similarity'])

# Add ylabel
plt.ylabel('Score', fontweight='bold')

# Add a legend
plt.legend()

# Show plot
plt.title('Comparison of Llama and ChatGPT')
plt.show()
```
